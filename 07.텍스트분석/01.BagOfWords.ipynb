### Bag Of Words

1. CountVectorizer

```python
text = 'My wife likes to watch baseball games and my daughter likes to watch baseball games too.'
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cvect = CountVectorizer()
output = cvect.fit_transform([text])
output.toarray()
```

```python
cvect.vocabulary_
```

- 불용어(Stopwords) 처리

```python
# 직접 지정
my_stopwords = ['to', 'and', 'my', 'too']
cvect = CountVectorizer(stop_words=my_stopwords)
print(cvect.fit_transform([text]).toarray())
print(cvect.vocabulary_)
```

```python
# Scikit-Learn에서 제공하는 불용어
cvect = CountVectorizer(stop_words='english')   # <- 'english'가 scikitlearn에서 제공하는 불용어
print(cvect.fit_transform([text]).toarray())
print(cvect.vocabulary_)
```

```python
# nltk에서 제공하는 불용어
from nltk.corpus import stopwords
sw = stopwords.words('english')
cvect = CountVectorizer(stop_words=sw)
print(cvect.fit_transform([text]).toarray())
print(cvect.vocabulary_)
```

- 인덱스에 해당하는 단어를 알려주는 함수

```python
voca = cvect.vocabulary_
for key, value in voca.items():
    print(key, value)
```

```python
def get_word(index, voca):
    for key, value in voca.items():
        if value == index:
            return key  
```

```python
get_word(4, cvect.vocabulary_)
```

2. TF-IDF (Terom Frequency - Inverse Document Frequency)

```python
text = ['My wife likes to watch baseball games and my daughter likes to watch baseball games too.',
        'My wife likes to play baseball.']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tvect = TfidfVectorizer(stop_words='english')
tvect.fit_transform(text).toarray()
```

```python
tvect.vocabulary_
```

```python
cvect = CountVectorizer(stop_words='english')
print(cvect.fit_transform(text).toarray())
print(cvect.vocabulary_)
```

3. N-gram

```python
text = ['I work at google.', 'I google at work']
cvect = CountVectorizer()
print(cvect.fit_transform(text).toarray())
print(cvect.vocabulary_)
```

```python
cvect = CountVectorizer(ngram_range=(1, 2))
print(cvect.fit_transform(text).toarray())
print(cvect.vocabulary_)
```
